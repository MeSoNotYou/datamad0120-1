To perform time series analysis on the data, we must change the index from a range index to a time series index. In the cell below, create a time series index using the `pd.date_range` function. Our next step is to decompose the time series and evaluate the patterns in the data. Load the `statsmodels.api` submodule and plot the decomposed plot of `userAcceleration.x`. Set `freq=60` in the `seasonal_decompose` function. Your graph should look like the one below.

![time series decomposition](../images/tsa_decompose.png)

# Challenge 2 - Modelling the Data

To model our data, we should look at a few assumptions. First, let's plot the `lag_plot` to detect any autocorrelation. Do this for `userAcceleration.x`

```python
pd.plotting.lag_plot(sensor['userAcceleration.x'])

```

Create a lag plot for `rotationRate.x`

```python
pd.plotting.lag_plot(sensor['rotationRate.x'])

```

What are your conclusions from both visualizations?

```python
# There does seem to be a degree of autocorrelation, but it does not completely explain the association between y(t) and y(t+1) (i.e. the correllation/regression is not perfect)
```

The next step will be to test both variables for stationarity. Perform the Augmented Dickey Fuller test on both variables below.

```python
from statsmodels.tsa.stattools import adfuller
```

```python
print(f'acceleration on x - p-value: {adfuller(acc)[1]}')
print(f'rotation rate on x - p-value: {adfuller(rot)[1]}')
```

What are your conclusions from this test?

```python
# for both variables, we can reject the null hypothesis that the variables are stationary at an alpha of 0.05, since both p-values are << 0.05.
# i.e. the variables are both not stationary
```

Finally, we'll create an ARMA model for `userAcceleration.x`. Load the `ARMA` function from `statsmodels`. The order of the model is (2, 1). Split the data to train and test. Use the last 10 observations as the test set and all other observations as the training set. 

```python
from statsmodels.tsa.arima_model import ARMA
```

```python
train = acc[:-10]
test = acc[-10:]
```

```python
model = ARMA(train, order=(2,1))
model_fit = model.fit()
```

```python
predicted_series =  model_fit.predict(start=len(train),
                    end=len(train)+len(test)-1, 
                    dynamic=False)
predicted_series
```

To compare our predictions with the observed data, we can compute the RMSE (Root Mean Squared Error) from the submodule `statsmodels.tools.eval_measures`. You can read more about this function [here](https://www.statsmodels.org/dev/generated/statsmodels.tools.eval_measures.rmse.html). Compute the RMSE for the last 10 rows of the data by comparing the observed and predicted data for the `userAcceleration.x` column.
